```python
import math
import pandas as pd
import numpy as np
def gradient_descent():
    m_curr=b_curr=0
    itr=1000
    lr_rate=0.001
    n=len(x)
    for i in range(itr):
        y_pr=m_curr*x+b_curr
        cost=(1/n)*sum((y-y_pr)**2)
        md=-(2/n)*sum(x*(y-y_pr))
        bd=-(2/n)*sum(y-y_pr)
        m_curr=m_curr-lr_rate*md
        b_curr=b_curr-lr_rate*bd
        print('m{}  ,b{}  ,cost{}'.format(m_curr,b_curr,cost))
        math.isclose(x,y,rel_tol=1e-20)
        
    
df=pd.read_csv("test_scores.csv")
x=np.array([i for i in df.math])
y=np.array([i for i in df.cs])
gradient_descent()
```

```python
import pandas as pd
from sklearn import linear_model
data=pd.read_csv('HR_comma_sep.csv')
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
encoded=ohe.fit_transform(data[['salary']])
encoded_cols=ohe.get_feature_names_out(['salary'])
dt=data.drop(['salary'],axis='columns')
encoded_df=pd.DataFrame(encoded,columns=encoded_cols)
df=pd.concat([dt,encoded_df],axis=1)
x=df.drop(['last_evaluation','Work_accident','Department','left'],axis='columns')
y=df['left']
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2)
model=linear_model.LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
pd.crosstab(data.salary,data.left).plot(kind='bar')
```

```python
from sklearn import datasets,linear_model
iris=datasets.load_iris()
```

```python
from matplotlib import pyplot as plt
plt.scatter(iris.data[:,0],iris.data[:,1],color='green')
plt.scatter(iris.data[:,2],iris.data[:,3],color='blue')
```

```python
dir(iris)
```

```python
iris.data
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(iris.data,iris.target,test_size=0.2)
```

```python
model=linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
plt.figure(figsize=(4,3))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
import pandas as pd 
from matplotlib import pyplot as plt
df=pd.read_csv('titanic.csv')
pd.crosstab(df.Survived,df.Pclass).plot(kind='barh')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex_n']=le.fit_transform(df['Sex'])
df=df.drop(['Sex','PassengerId','Name','SibSp','Parch','Cabin','Embarked','Ticket'],axis='columns')
df.head()
```

```python
x=df.drop(['Survived'],axis='columns')
y=df.Survived
from sklearn import tree
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2)
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,22.0,7.2500,1]])
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True)
```
